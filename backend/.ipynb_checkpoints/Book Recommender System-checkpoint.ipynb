## Popularity based recommender system

- Based on finding top 50 books with highest average rating 
- Consider only those books that are rated by atleast 250 users 

```python
 ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
ratings_num = ratings_with_name.groupby('bookTitle').count()['Book-Rating']
ratings_num.reset_index()
```

```python
no_ratings_for_each_books = ratings_with_name.groupby('bookTitle').count().reset_index()
```

```python
no_ratings_for_each_books
```

```python
no_ratings_for_each_books = ratings_with_name.groupby('bookTitle').count()['Book-Rating'].reset_index()
no_ratings_for_each_books.rename(columns={'Book-Rating':'numRatings'},inplace=True)
```

```python
no_ratings_for_each_books
```

```python
avg_ratings_for_each_books = ratings_with_name.groupby('bookTitle').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_ratings_for_each_books.rename(columns={'Book-Rating':'avgRatings'},inplace=True)
```

```python
avg_ratings_for_each_books
```

```python
popularity_df = no_ratings_for_each_books.merge(avg_ratings_for_each_books,on='bookTitle')
popularity_df
```

```python
top50 = popularity_df[popularity_df['numRatings'] >= 250].sort_values('avgRatings',ascending=False).head(50)
```

```python
top50
```

```python
top50 = top50.merge(books,on='bookTitle')
top50
```

```python
# 196 rows because one book have mutilple isbn's so dropping duplicates
```

```python
top50 = top50.drop_duplicates('bookTitle')[['bookTitle','bookAuthor','imageUrlM','numRatings','avgRatings']]
```

```python
top50.shape
```

```python
top50
```

```python
top50_book_info = {}

for _, row in top50.iterrows():
    title = row['bookTitle']
    author = row['bookAuthor']
    image = row['imageUrlM']
    
    if pd.notnull(title):
        top50_book_info[title] = {
            'author': author if pd.notnull(author) else "Unknown Author",
            'image': image if pd.notnull(image) else "https://drive.google.com/file/d/1F1OKkukD515raR37k9QkvPC6j7PGvMWJ/view?usp=sharing"
        }
```

```python
top50_book_info
```

```python
print(type(top50_book_info))
print(list(top50_book_info.items())[:3])  # Shows 3 entries
```

```python
import pickle

with open('top50_book_info.pkl', 'wb') as f:
    pickle.dump(top50_book_info, f)
```

## Collaborative Filtering Based Recommender System

based on books and ratings
- for an intelligent book recommender we must consider only those users who have rated atleast 200 books so only the vote of an experienced book reader is considered (knowledgable users)
- also only those books are considered whoch are rated by atleast 50 user, so very unpopular books are not considered (popular books)

(200 , 50) is the value that gave good results 

```python
# filtering
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
knowledgable_users = x[x].index
```

```python
knowledgable_users
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(knowledgable_users)]
filtered_ratings
```

```python
y = filtered_ratings.groupby('bookTitle').count()['Book-Rating'] >= 50
y[y]
```

```python
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['bookTitle'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='bookTitle',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

- every book e.g. book1 = [ ,  , ,........,810 numbers] is a vector in a 810 dimensional space
- now we can find distance of a vector from all others and find similar books based on this distance

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
similarity_score[0]
```

```python
# def recommend(book_name):
#     #index_fetch
#     index = np.where(pt.index == book_name)[0][0]
#     similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse=True)[1:9]

#     for i in similar_items:
#         print(pt.index[i[0]])
```

```python
def recommend(book_name):
    book_name = book_name.lower()

    # Convert index to lowercase for comparison
    lower_index = pt.index.str.lower()

    # Find all matches that contain the substring
    matches = [i for i, title in enumerate(lower_index) if book_name in title]

    if not matches:
        print("Book not found.")
        return

    # Use the first match
    index = matches[0]
    original_title = pt.index[index]
    print(f"\n Showing recommendations for: {original_title}\n")

    similar_items = sorted(
        list(enumerate(similarity_score[index])),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    for i in similar_items:
        title = pt.index[i[0]]
        author = ratings_with_name.get(title, {}).get('bookAuthor', 'Unknown Author')
        image = ratings_with_name.get(title, {}).get('imageUrlM', 'No image available')
        
        print(f"Title: {title}")
        print(f"Author: {author}")
        print(f"Image: {image}")
        print("-" * 60)
```

```python
recommend('harry potter')
```
